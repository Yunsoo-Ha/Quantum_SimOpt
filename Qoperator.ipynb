# A_operator with infinitely many shots

```python
q = QuantumRegister(3, 'q')
c = ClassicalRegister(1, 'c')

cost = 0.001
qo = QuantumCircuit(q, c)
```

```python
qo.ry(math.pi/2,q[0])
qo.ry(math.pi*2/3,q[1])
qo.cx(q[0],q[1])

qo.barrier(q[0:3])

qo.cry(4*1*np.sqrt(cost),q[0],q[2])
qo.cry(2*1*np.sqrt(cost),q[1],q[2])

qo.measure(q[2],c[0])
```

```python
qo.draw(output="mpl")
```

```python
backend = BasicAer.get_backend('qasm_simulator')
shots = 10000
results = execute(qo, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

```python
probability = answer['1']/shots
Approximate_answer = probability/cost
```

```python
print('Exact value:    \t%.4f' % exact_answer)
print('Estimated value:\t%.4f' % Approximate_answer)
```

# One Q operator

```python
def qoperator():
    ## Q_operator

    # Uf operator (constant for amplitude estimation)
    qo.barrier(q[0:3])

    qo.z(q[2])

    # A_inverse operator
    qo.cry(-2*1*np.sqrt(cost),q[1],q[2])
    qo.cry(-4*1*np.sqrt(cost),q[0],q[2])

    qo.barrier(q[0:3])

    qo.cx(q[0],q[1])
    qo.ry(-math.pi*2/3,q[1])
    qo.ry(-math.pi/2,q[0])

    # Us operator
    qo.barrier(q[0:3])

    qo.x(q[0:3])
    qo.h(q[2])
    qo.ccx(q[0],q[1],q[2])
    qo.h(q[2])
    qo.x(q[0:3])

    # A operator
    qo.barrier(q[0:3])

    qo.ry(math.pi/2,q[0])
    qo.ry(math.pi*2/3,q[1])
    qo.cx(q[0],q[1])

    qo.barrier(q[0:3])

    qo.cry(4*1*np.sqrt(cost),q[0],q[2])
    qo.cry(2*1*np.sqrt(cost),q[1],q[2])
```

```python
num_q = 1

q = QuantumRegister(3, 'q')
c = ClassicalRegister(1, 'c')

cost = 0.001
qo = QuantumCircuit(q, c)

qo.ry(math.pi/2,q[0])
qo.ry(math.pi*2/3,q[1])
qo.cx(q[0],q[1])

qo.barrier(q[0:3])

qo.cry(4*1*np.sqrt(cost),q[0],q[2])
qo.cry(2*1*np.sqrt(cost),q[1],q[2])

for x in range(0, num_q):
    qoperator()

qo.measure(q[2],c[0])
```

```python
qo.draw(output="mpl")
```

```python
backend = BasicAer.get_backend('qasm_simulator')
shots = 50000
results = execute(qo, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

```python
probability = answer['1']/shots
K = np.arcsin(np.sqrt(probability)) # theta*(2k+1)
k = K/(2*num_q+1) # theta
t = np.sin(k)
Approximate_answer = t**2/cost
```

```python
print('Exact value:    \t%.4f' % exact_answer)
print('Estimated value:\t%.4f' % Approximate_answer)
```
